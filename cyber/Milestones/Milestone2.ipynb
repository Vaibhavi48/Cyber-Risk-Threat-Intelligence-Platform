Port & Service Analysis

```python
import xml.etree.ElementTree as ET
tree = ET.parse("investigation.xml")
root = tree.getroot()
```

```python
for host in root.findall("host"):
    for port in host.findall(".//port"):
        port_id = port.get("portid")
        service = port.find("service")

        if service is not None:
            service_name = service.get("name")
            print(port_id, ":", service_name)
```

```python
for host in root.findall("host"):
    for port in host.findall(".//port"):
        service = port.find("service")

        if service is not None:
            name = service.get("name")
            version = service.get("version")

            print("Service:", name, "| Version:", version)
```

```python
risky_services = ["ssh", "ftp", "mysql", "smb"]

for host in root.findall("host"):
    for port in host.findall(".//port"):
        service = port.find("service")

        if service is not None:
            name = service.get("name")

            if name in risky_services:
                print("Sensitive Service Detected:", name)
```

Vulnerability Reasoning

1. Port 22 – SSH:
This port is used for remote login. If it is open to the public, attackers may try to guess passwords and break into the system.

2. Port 80 – HTTP:
This port is used for websites. Since it is not encrypted, data sent through it can be seen by others. If the website software is old, it may have security weaknesses.

3. Port 53 – DNS:
This port helps convert domain names into IP addresses. If not properly protected, attackers may misuse it to gather information or overload the server.

4. Port 9929 – nping-echo:
This is usually a testing service. If left open, it may give information about the network that could help attackers understand the system better.

5. Port 31337 – tcpwrapped:
This means the port is protected by some security control, like a firewall. Even though it is protected, it still shows that something is running behind it.
